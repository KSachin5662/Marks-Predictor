```python
#Feature Selection
new_data = pd.DataFrame(df,columns=["mid1","mid2","f_result"])
```

```python
new_data.corr(method='spearman')
```

```python
new_data.describe()
```

```python
x = pd.DataFrame(new_data,columns=['mid1','mid2'])
y = pd.DataFrame(new_data,columns=["f_result"])
```

```python
x.describe()
```

```python
x
```

```python
x['mid1'] = x['mid1']/2

x['mid1']
```

```python
x['mid2'] = x['mid2']/2
x['mid2']
```

```python
y['f_result'] = y['f_result']/2
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.1, random_state = 0)
```

```python
x.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
#Fitting
regression.fit(x_train,y_train)
```

```python
x_test
```

```python
y_pred = regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("r2_score =",r2_score(y_pred,y_test))
print("mean_squared_error =",mean_squared_error(y_pred,y_test))
```
